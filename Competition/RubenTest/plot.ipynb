```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")

stocks = ["AAA", "BBB", "CCC", "DDD", "ETF", "IND"]

for a in stocks:
    for b in stocks:
        plt.figure()
        plt.title(a + "/" + b)
        plt.scatter(df[a].values, df[b].values)
```
